# GLASSDOOR DATA JOBS

In this project, we're set to embark on a detailed analysis of Glassdoor data, aiming to draw a comparison between job listings for Data Science and Data Analyst roles. Our exploration will be structured in three key stages, each offering unique insights into the job landscape of these pivotal tech roles:

1. **Company Type Distribution:** We'll kick things off by dissecting the company types that are offering these roles. Be it private firms, public companies, or startups, we'll see where Data Scientists and Data Analysts are in demand. This phase will paint a clear picture of the corporate landscape, showing us the diversity and nature of organizations keen on harnessing data expertise.
2. **Sector-wise Job Distribution**: Moving on, we'll investigate how these job roles are distributed across various sectors. This analysis will help us pinpoint which industries are heavily leveraging Data Science and Data Analyst talents. It's an intriguing way to see how data-driven decision-making is permeating different fields.
3. **Company Ratings Comparison:** Our final analytical stride will involve a deep dive into company ratings. We'll compare firms offering Data Science and Data Analyst positions, analyzing how they stack up against each other in terms of career opportunities, compensation and benefits, culture and values, senior management, and work-life balance. This comprehensive comparison will offer a rounded view of the working environment and growth prospects within these roles.

Visualizations will play a key role in our analysis, with sky blue graphics representing Data Science jobs and light green for Data Analyst roles, ensuring that our findings are not only insightful but also visually compelling.

Get ready for a detailed and enlightening journey as we dissect and compare the intricate elements of these crucial roles in the technology sector.

## Data Scientist and Data Analyst Jobs

```python
data_scientist_jobs = df[df['job_title'].str.contains('Data Scientist|Data Science', case=False, na=False)]

data_scientist_jobs['job_title'].head(20)
```

```python
len(data_scientist_jobs)
```

```python
data_analyst_jobs = df[df['job_title'].str.contains('Data Analyst|Data Analytics|Analytics', case=False, na=False)]
data_analyst_jobs['job_title'].head(20)
```

```python
len(data_analyst_jobs)
```

```python
ds_employment_type_counts = data_scientist_jobs['employment_type'].value_counts()

plt.figure(figsize=(10, 6))
ds_employment_type_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Company Types for Data Scientist Jobs')
plt.xlabel('Company Type')
plt.ylabel('Number of Jobs')
plt.xticks(rotation=45)
plt.show()
```

```python
da_employment_type_counts = data_analyst_jobs['employment_type'].value_counts()

plt.figure(figsize=(10, 6))
da_employment_type_counts.plot(kind='bar', color='lightgreen')
plt.title('Distribution of Company Types for Data Scientist Jobs')
plt.xlabel('Company Type')
plt.ylabel('Number of Jobs')
plt.xticks(rotation=45)
plt.show()
```

```python
ds_valid_sectors = data_scientist_jobs[data_scientist_jobs['sector'] != '--']
ds_sectors = ds_valid_sectors['sector'].value_counts().head(5)

plt.figure(figsize=(10, 6))
ds_sectors.plot(kind='bar', color='skyblue')
plt.title('Distribution of Sector for Data Scientist Jobs')
plt.xlabel('Sector')
plt.ylabel('Number of Jobs')
plt.xticks(rotation=45)
plt.show()
```

```python
da_valid_sectors = data_analyst_jobs[data_analyst_jobs['sector'] != '--']
da_sectors = da_valid_sectors['sector'].value_counts().head(5)

plt.figure(figsize=(10, 6))
da_sectors.plot(kind='bar', color='lightgreen')
plt.title('Distribution of Sector for Data Analyst Jobs')
plt.xlabel('Sector')
plt.ylabel('Number of Jobs')
plt.xticks(rotation=45)
plt.show()
```

```python
ds_averages = data_scientist_jobs[['career_opportunities_rating', 'comp_and_benefits_rating', 
                                   'culture_and_values_rating', 'senior_management_rating', 
                                   'work_life_balance_rating']].mean()

da_averages = data_analyst_jobs[['career_opportunities_rating', 'comp_and_benefits_rating', 
                                 'culture_and_values_rating', 'senior_management_rating', 
                                 'work_life_balance_rating']].mean()

ratings_comparison = pd.DataFrame({'Data Scientist': ds_averages,
                                   'Data Analyst': da_averages})

ratings_comparison.plot(kind='bar', figsize=(10, 6))
plt.title('Comparison of Ratings between Data Scientists and Data Analysts Jobs')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
